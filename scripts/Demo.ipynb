```python
from pymldb import Connection

mldb = Connection("http://localhost:8080")
```

```python
url = "https://www.tensorflow.org/versions/r0.7/images/grace_hopper.jpg"

mldb.query("SELECT flatten(inception({url: '%s'})[softmax]) as *" % url)
```

```python
mldb.query("""
    SELECT scores.pred as score
    FROM transpose(
        (
            SELECT flatten(inception({url: '%s'})[softmax]) as *
            NAMED 'pred'
        )
    ) AS scores
""" % url)
```

```python
mldb.query("SELECT * FROM imagenet_labels limit 10")
```

```python
mldb.query("""
    SELECT scores.pred as score
    NAMED imagenet_labels.label
    FROM transpose(
        (
            SELECT flatten(inception({url: '%s'})[softmax]) as *
            NAMED 'pred'
        )
    ) AS scores

    LEFT JOIN imagenet_labels ON
        imagenet_labels.rowName() = scores.rowName()

    ORDER BY score DESC
    LIMIT 10
""" % url)
```
